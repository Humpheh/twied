```python
from pymongo import MongoClient

# connect to mongodb
client = MongoClient("localhost", 27018)
db = client["twitter"]

# get the tweet cursor
cursor = db.tweets.find({'entities.urls': { '$exists': True, '$not': {'$size': 0}}}, {'geo': 1, 'entities.urls': 1})

count = 0
docs = []
for d in cursor:
    docs.append(d)
    count += 1
    if count % 5000 == 0:
        print("Completed:", count)
        
print("Done")
```

```python
docs[0]
```

```python
from urllib.parse import urlparse
from collections import defaultdict

counts = defaultdict(int)
for d in docs:
    for e in d['entities']['urls']:
        domain = urlparse(e['expanded_url']).netloc
        counts[domain] += 1
```

```python
sc = sorted(counts.items(), key=lambda x: x[1], reverse=True)
sc
```
